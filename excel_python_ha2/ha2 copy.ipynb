```python
def plot_histograms(data, columns, bins=30):
    """
    Plots simple histograms for specified columns in a dataframe.

    Parameters:
    ----------
    data : pandas.DataFrame
        The dataframe containing the data to plot.
    columns : list of str
        List of column names to generate histograms for.
    bins : int, optional
        The number of bins for the histogram (default is 30).

    Returns:
    -------
    None
        Displays the histogram plots for the specified columns.
    """
    for column in columns:
        # Plot the histogram for the current column
        data[column].dropna().plot(kind='hist', bins=bins, title=f"Histogram for {column}", figsize=(8, 5))
        
        # Add labels
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True, alpha=0.3)
        
        # Show the plot
        plt.show()
```

```python
# Example: Plot for all goal columns
goal_columns = [f"Goal {i}" for i in range(1, 18)]
plot_histograms(df4, goal_columns)

```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_correlation_matrix(data, columns):
    """
    Plots a correlation matrix heatmap with displayed values for specified columns in a dataframe.

    Parameters:
    ----------
    data : pandas.DataFrame
        The dataframe containing the data to analyze.
    columns : list of str
        List of column names to include in the correlation matrix.

    Returns:
    -------
    None
        Displays the correlation matrix heatmap with values.
    """
    # Compute the correlation matrix
    correlation_matrix = data[columns].corr()

    # Plot the heatmap
    plt.figure(figsize=(10, 8))
    plt.matshow(correlation_matrix, fignum=1, cmap='coolwarm')

    # Add labels and title
    plt.xticks(range(len(columns)), columns, rotation=90)
    plt.yticks(range(len(columns)), columns)
    plt.colorbar(label="Correlation Coefficient")
    plt.title("Correlation Matrix", pad=20)

    # Display the correlation values on the heatmap
    for (i, j), value in np.ndenumerate(correlation_matrix):
        plt.text(j, i, f"{value:.2f}", ha='center', va='center', color="black" if abs(value) < 0.7 else "white")

    # Show the heatmap
    plt.show()
```

```python
# Example: Plot for all goal columns
goal_columns = [f"Goal {i}" for i in range(1, 18)]
plot_correlation_matrix(df4, goal_columns)
```

```python
def display_top_correlations(data, columns, top_percent=10):
    """
    Displays the top N% least and most correlated unique pairs of columns in a dataframe.

    Parameters:
    ----------
    data : pandas.DataFrame
        The dataframe containing the data to analyze.
    columns : list of str
        List of column names to include in the correlation matrix.
    top_percent : int, optional
        The percentage of top correlations to display (default is 10%).

    Returns:
    -------
    None
        Prints the top N% least and most correlated unique pairs.
    """
    # Compute the correlation matrix
    correlation_matrix = data[columns].corr()

    # Extract the upper triangle of the matrix without the diagonal
    correlation_pairs = correlation_matrix.where(
        np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
    ).stack()

    # Sort the correlation values
    sorted_correlations = correlation_pairs.sort_values()

    # Determine the number of pairs to display
    total_pairs = len(sorted_correlations)
    top_n = max(1, int(total_pairs * (top_percent / 100)))  # Ensure at least one pair is selected

    # Get the least and most correlated pairs
    least_correlated = sorted_correlations[:top_n]
    most_correlated = sorted_correlations[-top_n:]

    # Display the results
    print("Top Least Correlated Pairs:")
    print(least_correlated)

    print("\nTop Most Correlated Pairs:")
    print(most_correlated)
```

```python
# Example: Find top 10% least and most correlated pairs for goal columns
goal_columns = [f"Goal {i}" for i in range(1, 18)]
display_top_correlations(df4, goal_columns, top_percent=3)
```
