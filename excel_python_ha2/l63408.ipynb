### 1: Import Libraries:
First, we import the Pandas library for data manipulation and visulisation. 

For the data visualisation we import matplotlib library as well.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 2: Load and Clean Data:
Next, we load the dataset using Pandas (`pd.read_csv()`), inspect first and last few rows using `head()` and `tail()`. 

Since the first seven rows did not include relevant data we exclude them by taking into account only the rows starting from the row 8 using `df[7:]`.


```python
df = pd.read_csv('CountryWiseGoal.csv')

df.head(10)
```

```python
df.tail()
```

```python
df2 = df[7:]
df2
```

### 3: Inspect DataFrame Structure
We use the `info()` method to see the structure of the dataframe `df2`.

We can see that the dataframe contains 251 rows and 19 columns. However, the names of the columns do not seemt to be correct.

Importantly, all columns have non-null values.

Finally, the data type `object` might not be of use to us; therefore, we will convert it to a suitable data format later.


```python
df2.info()
```

### 4: Rename Columns:
We rename the columns of the DataFrame:

Since the first row (at index 0) of `df2` contains the correct column names, we assign these values to the dataframe columns using `df2.columns = df2.iloc[0]`.

Then we create a copy of the dataframe - `df3 = df2.copy()`

```python
df2.columns = df2.iloc[0]
df3 = df2.copy()
df3
```

### 5: Remove the first row:

Since the first row of `df3` was used as column names, we exclude it from the data using `df3[1:]` and create a copy of the new dataframe `df4`.


```python
df4 = df3[1:].copy()
df4
```

### 6: Reset the Index:

Since we deleted first few rows, we use `reset_index()` with `drop=True` and `inplace=True` so that we drop the old index and assign the new one directly to `df4`.

```python
df4.reset_index(drop=True, inplace=True)
df4
```

### 7: Remove index column name:
We remove the current name of the index column ('7') using `df4.columns.name = None`.


```python
df4.columns.name = None
df4
```

### 8: Clean Goal Columns:
Next, we strip the `%` symbol from the goal columns and convert the cleaned values to `float` so that we can work with them further.
We apply this to all columns except the first two using a loop.


```python
#remove '%' and convert numeric columns to float
goal_columns = df4.columns[2:]  #excluding the firs two columns
for column in goal_columns:
    df4[column] = df4[column].str.rstrip('%').astype(float)

df4
```

### 9: Data Summary:
Using the `.describe()` method, we generate summary statistics for the goal columns as follows:


```python
summary_stats = df4.describe()
summary_stats
```

### 10: Plot Histograms
Next we create and use a function, generating a histogram (distribution of values) for each of the goal columns in our dataframe:


```python
def plot_histograms(data, columns, bins=30):
    """
    Plots histograms for specified columns in a dataframe.

    Parameters:
    ----------
    data : pandas.DataFrame
        The dataframe containing the data to plot.
    columns : list of str
        List of column names to generate histograms for.
    bins : int, optional
        The number of bins for the histogram (default is 30).

    Returns:
    -------
    None
        Displays the histogram plots for the specified columns.
    """
    for column in columns:
        data[column].dropna().plot(kind='hist', bins=bins, title=f"Histogram for {column}", figsize=(8, 5))
        
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True, alpha=0.3)
        
        plt.show()
```

```python
#plot for all goal columns
goal_columns = [f"Goal {i}" for i in range(1, 18)]
plot_histograms(df4, goal_columns)
```

### 11: Rename Columns:
Next, we assign new names to the goal columns to both the original dataframe and a summary statistics dataframe, using a list of new column names.

```python
#the new column names list:
new_column_names = [
    "GeoAreaCode",
    "GeoAreaName", 
    "No Poverty", 
    "Zero Hunger", 
    "Good Health and Well-Being", 
    "Quality Education", 
    "Gender Equality", 
    "Clean Water and Sanitation", 
    "Affordable and Clean Energy", 
    "Decent Work and Economic Growth", 
    "Industry, Innovation and Infrastructure", 
    "Reduced Inequalities", 
    "Sustainable Cities and Communities", 
    "Responsible Consumption and Production", 
    "Climate Action", 
    "Life below Water", 
    "Life on Land", 
    "Peace, Justice and Strong Institutions", 
    "Partnerships for the Goals"
]

#rename the columns
df4.columns = new_column_names
summary_stats.columns = new_column_names[2:] #starting from the 3rd name, since summary_stats has less columns
df4
```

```python
summary_stats
```

### 12: Write Data to Excel
Finally, we write the processed data to a single excel file. To the first sheet we upload the processed dataset and to the second sheet the summary statistics.

```python
with pd.ExcelWriter("processed_data.xlsx", engine="openpyxl") as writer:
    #write the data to the first sheet:
    df4.to_excel(writer, sheet_name="Processed Data", index=False)
    
    #write the summary statistics to the second sheet:
    summary_stats.to_excel(writer, sheet_name="Summary Statistics")
```
