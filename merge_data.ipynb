```python
import pandas as pd

# Load the JSON files
card_details = pd.read_json('products_singles_6(2).json')
price_data = pd.read_json('price_guide_6(3).json')

```

```python
print(card_details.columns)
print(price_data.columns)

```

```python

# Extract the nested data
card_details_flat = pd.json_normalize(card_details['products'])
price_data_flat = pd.json_normalize(price_data['priceGuides'])

# Display the extracted data to verify
card_details_flat.head()

```

```python
price_data_flat.head()
```

```python
# Merge on 'idProduct'
merged_data = pd.merge(card_details_flat, price_data_flat, on='idProduct', how='inner')

# Display the merged data
merged_data.head()
```
