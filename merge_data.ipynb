```python
import pandas as pd

# Load the JSON files
card_details = pd.read_json('products_singles_6(2).json')
price_data = pd.read_json('price_guide_6(3).json')

```

```python

# Extract the nested data
card_details_flat = pd.json_normalize(card_details['products'])
price_data_flat = pd.json_normalize(price_data['priceGuides'])

# Display the extracted data to verify
card_details_flat.head()

```

```python
# Merge on 'idProduct'
merged_data = pd.merge(card_details_flat, price_data_flat, on='idProduct', how='inner')

# Display the merged data
merged_data.head()
```

```python
import requests
from bs4 import BeautifulSoup

# Define the base URL for the page you want to scrape
url = "https://www.cardmarket.com/"  # Adjust for Pokémon or other categories

# Headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}

# Send a GET request
response = requests.get(url, headers=headers)
response_text = response.text
print(response_text)
# Check the response status
if response.status_code == 200:
    print("Request successful!")
    soup = BeautifulSoup(response.content, "html.parser")
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python

# Parse the content (update selectors based on the structure)
cards = soup.find_all("div", class_="row no-gutters")  # Example class for card rows (inspect the actual site)

for card in cards:
    try:
        name = card.find("a", class_="product-name").text.strip()
        price = card.find("span", class_="price").text.strip()
        print(f"Card Name: {name}, Price: {price}")
    except AttributeError:
        # Skip rows that don't match the expected structure
        continue
```
